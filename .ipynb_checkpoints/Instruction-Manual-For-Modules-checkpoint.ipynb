# Thermodynamics Module

The thermodynamic properties necessary for the treatment of heat transfer problem include:
* Density $\rho \text{ (kg/m$^3$)}$
* Specific heat at constant pressure $C_p \text{ (J/(kg.K)}$
* Dynamic viscosity $\mu \text{ (kg/(m.s))}$
* Thermal conductivity $k \text{ (W/(m.K))}$
* Kinematic viscosity $\nu \text{ (m$^2$/s)}$
* Thermal diffusivity $\alpha \text{ (m$^2$/s)}$
* Prandtl number $Pr$, ratio of kinematic viscosity to thermal diffusivity, dimensionless
* Thermal expansion coefficient $\beta \text{ (1/K)}$


The module <FONT FACE="courier" style="color:blue">from Libraries import thermodynamics as thermo </FONT> gives these thermodynamics properties for water, air, argon and krypton under <FONT style="color:red">$1\text{ atm}$</FONT> over a given range of temperature $T$ (dependent upon the fluid). Air properties are tabled between $-150\text{$^\circ$C}$ and $400\text{$^\circ$C}$, water between $274\text{K}$ and $373\text{K}$, Argon between $100\text{K}$ and $700\text{K}$ and Krypton between $150\text{K}$ and $750\text{K}$
<p class='alert alert-danger'>
    <b>When using the library thermodynamics, the input temperature must be in Kelvin</b>
</p>
Use the library's functions <FONT FACE="courier" style="color:blue">C2K, K2C, K2F </FONT> and <FONT FACE="courier" style="color:blue">F2K </FONT> to convert temperatures to Kelvin, Celsius or Fahrenheit. Thermodynamic properties are linearly interpolated from the two nearest temperature states.

## Thermal Resistance Module

The class thermal resistance is imported with <FONT FACE="courier" style="color:blue">from Libraries import HT_thermal_resistance as res</FONT>.

### Heat rates, rates per unit length or heat fluxes

Thermal resistances are derived from an analogy between heat transfer and electric circuit. Each mode of heat transfer can be represented by a thermal resistance, linking the heat flux $q''~(\mathrm{W/m^2}$), heat rate per unit length $q'~(\mathrm{W/m}$) or heat rate $q~(\mathrm{ W}$) to the temperature difference between two points ($T_a-T_b$, with $T_a>T_b$:

<p class='alert alert-danger'>
$$
q''=\frac{1}{R''}\left(T_a-T_b\right),\; q'=\frac{1}{R'}\left(T_a-T_b\right),\; q=\frac{1}{R}\left(T_a-T_b\right)
$$
</p>

The heat transfer rate, also called heat loss or heat gain, is analog to electric current, and  temperature to potential difference. 

### Thermal resistances

For 1D conduction in a solid or a fluid at rest, the expression of the thermal resistance depends upon the geometry of the system:
* **Plane wall** or planar conduction is declared by the function `R[i].cond_plane(k,L)` for calculation of heat flux ($W/m^2$) or `R[i].cond_plane(k,L,A)` for calculation or heat rate ($W$). `k` is the thermal conductivity, `L` the wall thickness and `A` is the cross sectional area of the wall (perpendicular to the flux).
<p class='alert alert-danger'>
$$
R_\text{th,cond}=\frac{L}{kA}
$$
</p>

* **Cylindrical shell**: `R[i].cond_cylinder(k,ra,rb)` or `R[i].cond_cylinder(k,ra,rb,L=Lpipe ,)` or `R[i].cond_cylinder(k,ra,rb,angle)` or `R[i].cond_cylinder(k,ra,rb,L=Lpipe ,angle = angle_pipe)` . The first computes $R'$ for a unit length of the pipe, the second $R$ for the whole pipe, the third $R'$ for a cylindrical shell with a specific angular dimenion (i.e. smaller than $2\pi$) and the fourth is the same as the third but for the whole length of the shell. `ra` and `rb` are the internal and external radii of the shell.
<p class='alert alert-danger'>
$$
R_\text{th,cond}=\cfrac{\ln\left(\cfrac{r_b}{r_a}\right)}{2\pi kL}
$$
</p>

* **Spherical shell**: `R[i].cond_sphere(k,ra,rb)` or `R[i].cond_sphere(k,ra,rb,scale)`. Here scale is a value <= 1., 1. by default for a full spherical shell, less than one for a fraction shell. For instance, a cornea is about 1/3 of spherical shell, therefore scale=1./3.
<p class='alert alert-danger'>
$$
R_\text{th,cond}=\cfrac{\cfrac{1}{r_a}-\cfrac{1}{r_b}}{4\pi k}
$$
</p>

* **Contact resistance**: `R[i].contact(Rc)` or `R[i].contact(Rc,A)`. `A` is optional and `A=1.0` by default when calculating heat flux. `Rc`, the contact resistance is obtained from tables based on the materials involved and their respective roughness.

* **Convection**: `R[i].convection(h)` or `R[i].convection(h, A)`. `A` is optional and `A=1.0` by default when calculating heat flux. `h` is the convection heat transfer coefficient ($W/(m^2.K)$ obtained from correlations specific to the geometry and flow parameters of the problem.

<p class='alert alert-danger'>
$$
R_\text{th,conv}=\frac{1}{hA}
$$
</p>

* **Radiation**: `R[i].radiation(eps,T_s,T_sur)` and `R[i].radiation(eps,T_s,T_sur,A)`. `A` is optional and `A=1.0` by default when calculating heat flux. `eps` is the permissivity of the material, `T_s` the surface temperature and `T_sur` is the temperature of the distant surface emitting radiation. **Both temperatures must be in Kelvin**. `R[i].h` outputs the radiation heat transfer coefficient $h_r$:

<p class='alert alert-danger'>
$$
R_\text{th,rad}=\frac{1}{h_rA}\, \text{ with } h_r = \varepsilon\sigma(T_s+T_\text{sur})(T_s^2+T_\text{sur}^2)
$$
<b> To calculate $h_r$, temperatures must be in Kelvin</b>
</p>

Like a thermal circuit, $n$ thermal resistances in serial are equivalent to one total resistance. If the $n$ resistances are stored in an array $R[i]$, the total resistance of resistances in series starting at $m$ and finishing at $n$ is 
<FONT FACE="courier" style="color:blue">serial_sum(R,m,n)</FONT>
<p class='alert alert-danger'>
$$
R_\text{th,tot}=\sum_{i=m}^{n}R[i]
$$
</p>
Similarly, thermal resistances in parallel between $n$ and $m$ reduce to
<FONT FACE="courier" style="color:blue">parallel_sum(R,m,n)</FONT>
<p class='alert alert-danger'>
$$
R_\text{th,tot}=\left(\sum_{i=m}^{n}\cfrac{1}{R[i]}\right)^{-1}
$$
</p>


The first step is to define a list of thermal resistance, and give each resistance its own label.

Here we use the class Resistance in the module HT_thermal_resistance. The first call requires the label (in Latex format with `"$ $"`) and the dimensional unit of the associated flux or heat rate: `'W/m^2'`, `'W/m'`, `'W'`. To help you remember what units you are dealing with it is strongly advised to use the list name `R` for thermal resistances defining a heat rate, `Rp` for $R'$ (for heat rate per unit length with label `"$R'_{cond}$"`), or `Rpp` for $R''$ for heat flux with label `"$R''_{cond}$"` Note that the double prime in LaTeX (i) requires the use of double quotes `"` to define the beginning and end of the string, (ii) is `''` like we spell it: two successive primes. 


### Example 

Consider a heated fluid flowing in a pipe, resulting in a convection coefficient $h_i$. The pipe's inner, outer radii, and thermal conductivity are known. The outer surface of the pipe experiences convection with the ambient fluid. The convection coefficient with the outside if $h_o$, and the outside ambient fluid is a temperature $T_\infty$ (far away from the outer surface of the pipe). The inner and outer surfaces of the pipe are at temperature $T_{s,i}$, and $T_{s,o}$, respectively. In a given cross-section, the average heated fluid temperature is $T_{m,i}$. You are asked to define the total and individual heat transfers for this cross-section, here you are dealing with heat rate per unit length. 

Even though the heat transfer by radiation between the outer surface and the outside ambient fluid is defined, we assume it to be negligible for now.


```python
from Libraries import HT_thermal_resistance as res

Rp = []
Rp.append(res.Resistance("$R'_{conv,i}$",'W/m')) # Rth[0]
Rp.append(res.Resistance("$R'_{cond,pipe}$",'W/m')) #Rth[1]
Rp.append(res.Resistance("$R'_{conv,o}$",'W/m')) #Rth[2]
Rp.append(res.Resistance("$R'_{rad,o}$",'W/m')) #Rth[3]

print('number of resistances %2i' %len(Rp))
```

Typically the next step is to draw the thermal circuit using the module `schemdraw` already imported in the first cell:
<FONT FACE="courier" style="color:blue">
import schemdraw as schem
</FONT>
and
<FONT FACE="courier" style="color:blue">
import schemdraw.elements as elm
</FONT>


Let's start with the first thermal resistance `R'[0]`, which models internal convection between an average bulk temperature $T_{m,i}$ and the temperature at the inner surface of the pipe $T_{s,i}$. The sketch starts with a dot element, defining the temperature $T_{m,i}$ written below the dot with the instruction `'bottom'`. Labels can be located at the `'top'` (default), '`bottom'`, `'right'`, or `'left'` of the element they are attached to. Omission of this instruction as an argument of the function `.label()` places the label on the default location `'top'`.

The second element is the resistor, placed to the `.right()` of the previous element. `Resistor` and `Line` elements can be oriented `.right()` (default), `.left()`, `.up()` or `.down()`. Omission of the direction orients the `Resistor` or `Line` element in the default direction. 

The third element is the end temperature $T_{s,i}$ shown as a `Dot`.

The next two cells show two different variations of the same circuit, one with specific label locations and element orientation, and one with the default label locations and element orientation.

```python
import schemdraw as schem
import schemdraw.elements as elm
with schem.Drawing() as d:
    d += elm.Dot().label("$T_{m,i}$",'bottom')
    d += elm.Resistor().up().label(Rp[0].name,'top')
    d += elm.Dot().label("$T_{s,i}$",'right')
```

```python
import schemdraw as schem
import schemdraw.elements as elm
with schem.Drawing() as d:
    d += elm.Dot().label("$T_{m,i}$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
```

Now the last two resistances are added using default label locations and element orientation. Note how `schemdraw` connect each element with the previous one, going from left to right, if the direction and location of the elements are not defined.

```python
import schemdraw as schem
import schemdraw.elements as elm
with schem.Drawing() as d:
    d += elm.Dot().label("$T_{m,i}$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += elm.Resistor().label(Rp[1].name)
    d += elm.Dot().label("$T_{s,o}$")
    d += elm.Resistor().label(Rp[2].name)
    d += elm.Dot().label("$T_{\infty}$")
```

### Showing heat fluxes, heat rates per unit length or, heat rates

An accurate description of the heat transfer system includes the heat fluxes, heat rates per unit length, or heat rates involved and their direction. The element `Line` can be supplemented with an arrow, and placed at any location relative to the first element of the circuit. The coordinates of the first element of the circuit are $(0,0)$. The default length of the `Line` or `Resistor` elements is $3$. A total heat rate is added to the left of the thermal circuit, starting at $(-2.5,0)$ and with a length of $2.25$. The arrow is an argument of the element `Line(arrow='->')`, which can be changed to the opposite direction `Line(arrow='<-')`.

```python
import schemdraw as schem
import schemdraw.elements as elm
with schem.Drawing() as d:
    d += elm.Dot().label("$T_{m,i}$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += elm.Resistor().label(Rp[1].name)
    d += elm.Dot().label("$T_{s,o}$")
    d += elm.Resistor().label(Rp[2].name)
    d += elm.Dot().label("$T_{\infty}$")
    d += elm.Line(arrow='->').right().at((-2.5,0)).length(2.25).label("$q'$").color('red')
```

Now the heat rates for each resistance are added underneath their respective resistor.

```python
with schem.Drawing() as d:
    d += elm.Dot().label("$T_m$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += elm.Resistor().label(Rp[1].name)
    d += elm.Dot().label("$T_{s,o}$")
    d += elm.Resistor().label(Rp[2].name)
    d += elm.Dot().label("$T_{\infty}$")
    d += elm.Line(arrow='->').right().at((-2.5,0)).length(2.25).label("$q'$").color('red')
    d += elm.Line(arrow='->').right().at((0.25,-1)).length(2.5).label("$q'_{conv,i}$",'bottom').color('blue')
    d += elm.Line(arrow='->').right().at((3.25,-1)).length(2.5).label("$q'_{cond}$",'bottom').color('orange')
    d += elm.Line(arrow='->').right().at((6.25,-1)).length(2.5).label("$q'_{conv,o}$",'bottom').color('green')
```


Assuming that radiation is no longer negligible, the surface energy balance on the outer surface is now

$$
q'_{cond}=q'_{conv,o}+q'_{rad,o}
$$
therefore requiring two resistors in parallel. First, the serial part of the circuit is sketched up to the outer surface:

```python
with schem.Drawing() as d:
    d += elm.Dot().label("$T_m$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += (R1:=elm.Resistor().label(Rp[1].name))
    d += elm.Dot().label("$T_{s,o}$")
```


Note that the last resistor has now a tag `R1` for outer surface. We can use this tag as a starting point as shown below.

To create the parallel circuit, the upper branch is defined as

```python
with schem.Drawing() as d:
    d += elm.Dot().label("$T_m$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += (R1 :=elm.Resistor().label(Rp[1].name))
    d += elm.Dot().label("$T_{s,o}$",'right')
    d += elm.Line().up().length(1.5)
    d += elm.Resistor().right().label(Rp[2].name)
    d += elm.Dot().label("$T_\infty$")
```

To sketch the lower branch, the next element needs to start at the `Dot` tagged OS

```python
with schem.Drawing() as d:
    d += elm.Dot().label("$T_m$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += (R1 :=elm.Resistor().label(Rp[1].name))
    d += elm.Dot().label("$T_{s,o}$",'right')
    d += elm.Line().up().length(1.5)
    d += elm.Resistor().right().label(Rp[2].name)
    d += elm.Dot().label("$T_\infty$")
    d += elm.Line().at(R1.end).down().length(1.5)
    d += elm.Resistor().right().label(Rp[3].name)
    d += elm.Dot().label("$T_{sur}$")
```

```python
with schem.Drawing() as d:
    d += elm.Dot().label("$T_m$")
    d += elm.Resistor().label(Rp[0].name)
    d += elm.Dot().label("$T_{s,i}$")
    d += (R1 :=elm.Resistor().label(Rp[1].name))
    d += elm.Dot().label("$T_{s,o}$",'right')
    d += elm.Line().up().length(1.5)
    d += elm.Resistor().right().label(Rp[2].name)
    d += elm.Dot().label("$T_\infty$")
    d += elm.Line().at(R1.end).down().length(1.5)
    d += elm.Resistor().right().label(Rp[3].name)
    d += elm.Dot().label("$T_{sur}$")
    d += elm.Line(arrow='->').right().at((-2.25,0)).length(2).label("$q'$").color('red')
    d += elm.Line(arrow='->').right().at((0.25,-0.5)).length(2.5).label("$q'_{conv,i}$",'bottom').color('blue')
    d += elm.Line(arrow='->').right().at((3.25,-0.5)).length(2.5).label("$q'_{cond}$",'bottom').color('orange')
    d += elm.Line(arrow='->').right().at((6.25,1)).length(2.5).label("$q'_{conv,o}$",'bottom').color('green')
    d += elm.Line(arrow='->').right().at((6.25,-2)).length(2.5).label("$q'_{rad,o}$",'bottom').color('violet')
```
